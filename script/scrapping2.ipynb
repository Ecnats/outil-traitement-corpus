```python
import csv
import requests
import lxml.html
import os

# Définir le chemin relatif du fichier de sortie
filename = '../data/histoire.csv'

# Initialiser un ensemble pour stocker les contenus déjà écrits
written_contents = set()

# S'assurer que le dossier parent 'data' existe
os.makedirs(os.path.dirname(filename), exist_ok=True)

with open(filename, 'a', newline='', encoding='utf-8') as csvfile:
    full_text = csv.writer(csvfile, delimiter='\t', quotechar='"', quoting=csv.QUOTE_ALL)

    # Écrire les en-têtes une seule fois, avant la boucle, seulement si le fichier est vide
    try:
        if os.stat(filename).st_size == 0:
            headers = ["links", "histoire"]
            full_text.writerow(headers)
    except FileNotFoundError:
        # Si le fichier n'existe pas, écrire les en-têtes
        headers = ["links", "histoire"]
        full_text.writerow(headers)

    response = requests.get('https://www.iletaitunehistoire.com/genres/contes-et-legendes')
    
    if response.status_code == 200:
        html_content = response.content
        tree = lxml.html.fromstring(html_content)
        
        # Trouver tous les liens <a>
        links = tree.xpath('//a')
        
        for link in links:
            href = link.get('href')

            # pour récupérer le résumé
            if href and href.startswith('/genres/contes-et-legendes/lire/') and href.endswith('#histoire'):
                full_link = 'https://www.iletaitunehistoire.com' + href
                
                # Effectuer une nouvelle requête HTTP vers l'URL du lien
                response_link = requests.get(full_link)
                
                if response_link.status_code == 200:
                    tree_link = lxml.html.fromstring(response_link.content)
                    
                    # Chemin XPath pour les paragraphes
                    path = '//div[@class="idocBlock"]'

                    # Extraire les paragraphes
                    paragraphs = tree_link.xpath(path)
                    
                    # Afficher le contenu des paragraphes
                    content = ' '.join([p.text_content() for p in paragraphs])

                    # Éviter les doublons
                    if content not in written_contents:
                        # Ajouter le contenu à l'ensemble pour éviter les doublons
                        written_contents.add(content)

                        # Écrire le contenu dans le fichier CSV
                        full_text.writerow([full_link, content])
                        
                        print(f"URL: {full_link}")
                        print(f"Content1:\n{content}\n")

                else:
                    print(f"Erreur lors de la récupération de la page {full_link}: {response_link.status_code}")
    else:
        print('Erreur lors de la récupération de la page principale')
```

```python
import csv
import requests
import lxml.html

# Définir le chemin relatif du fichier de sortie
filename = '../data/resume.csv'

# Lire le contenu du fichier CSV et stocker dans une variable (décommenter si nécessaire)
# with open(filename, 'r', newline='', encoding='utf-8') as readfile:
#     existing_contents = set(readfile.read().splitlines())

with open(filename, 'a', newline='', encoding='utf-8') as csvfile:
    full_text = csv.writer(csvfile, delimiter=' ')

    # Écrire les en-têtes une seule fois, avant la boucle, seulement si le fichier est vide
    try:
        if os.stat(filename).st_size == 0:
            headers = ["url", "resume"]
            full_text.writerow(headers)
    except FileNotFoundError:
        # Si le fichier n'existe pas, écrire les en-têtes
        headers = ["url", "resume"]
        full_text.writerow(headers)

    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.93 Safari/537.36'}

    response = requests.get('https://www.iletaitunehistoire.com')

    if response.status_code == 200:
        html_content = response.content
        tree = lxml.html.fromstring(html_content)

        # Trouver tous les liens <a>
        links = tree.xpath('//a')

        for link in links:
            href = link.get('href')

            # pour récupérer le résumé
            if href and href.startswith('/genres/contes-et-legendes'):
                full_link = 'https://www.iletaitunehistoire.com' + href

                # Effectuer une nouvelle requête HTTP vers l'URL du lien
                response_link = requests.get(full_link)

                if response_link.status_code == 200:
                    tree_link = lxml.html.fromstring(response_link.content)

                    # Chemin XPath pour les balises <h2>
                    div_elements = tree_link.xpath('//div[@class="idocAbstract"]')

                    # Afficher le contenu des divs idocAbstract
                    for div in div_elements:
                        # Extraire le contenu de la div après suppression des balises <h2>
                        content = div.text_content().strip()

                        # Écrire le contenu dans le fichier CSV
                        full_text.writerow([full_link, content])
                        print(f"Content:\n{content}\n")
                else:
                    print(f"Erreur lors de la récupération de la page {full_link}: {response_link.status_code}")
    else:
        print('Erreur lors de la récupération de la page principale')
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Définir le chemin relatif du fichier de sortie
output_file = '../data/resumes.csv'

# Récupérer le contenu de l'URL
url = "https://www.iletaitunehistoire.com/genres/contes-et-legendes"
response = requests.get(url)
response.raise_for_status()  # S'assurer que la requête a réussi

# Analyser le contenu HTML avec BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Trouver toutes les divs story-card-container
story_containers = soup.find_all('div', class_='col-12 col-sm-6 story-card-container')

# Liste pour stocker les données extraites
data = []

# Itérer sur chaque container et extraire les informations nécessaires
for container in story_containers:
    # Trouver la balise a story-link et obtenir l'attribut href
    story_link = container.find('a', class_='story-link')
    link = story_link.get('href') if story_link else 'N/A'
    
    # Trouver la div idocAbstract et obtenir son contenu
    idoc_abstract = container.find('div', class_='idocAbstract')
    content = idoc_abstract.text.strip() if idoc_abstract else 'N/A'
    
    # Ajouter les données à la liste
    data.append({'links': "https://www.iletaitunehistoire.com" + link, 'resumes': content})

# Convertir la liste en DataFrame
df = pd.DataFrame(data)

# Enregistrer la DataFrame dans un fichier CSV
df.to_csv(output_file, index=False, encoding='utf-8', sep='\t')

print(f"Les données ont été sauvegardées dans {output_file}")
```

```python
import pandas as pd

# Charger les deux fichiers CSV
csv1 = '../data/histoire.csv'
csv2 = '../data/resumes.csv'

# Fonction pour lire le CSV tout en ignorant les lignes mal formées
def read_csv_safely(file_path, sep='\t'):
    try:
        return pd.read_csv(file_path, sep=sep)
    except pd.errors.ParserError as e:
        print(f"Erreur de parsing pour le fichier {file_path}: {e}")
        return pd.DataFrame()  # Retourne un DataFrame vide en cas d'erreur

# Lire les fichiers CSV dans des DataFrames
df1 = read_csv_safely(csv1)
df2 = read_csv_safely(csv2)

# Vérifier les premières lignes des DataFrames pour s'assurer qu'ils sont correctement lus
print("DataFrame df1:")
print(df1.head())
print("\nDataFrame df2:")
print(df2.head())

# Fusionner les DataFrames sur la colonne URL commune
merged_df = pd.merge(df1, df2, on='links', how='inner')

# Supprimer les lignes dupliquées
merged_df = merged_df.drop_duplicates()

# Sélectionner les colonnes requises pour le fichier CSV final
final_df = merged_df[['histoire', 'resumes']]

# Ajouter une colonne 'id' qui s'incrémente de 1 pour chaque ligne
final_df.insert(0, 'id', range(1, len(final_df) + 1))

# Enregistrer le DataFrame fusionné dans un nouveau fichier CSV
output_file = '../data/histoires_resumes.csv'
final_df.to_csv(output_file, index=False, encoding='utf-8')

print(f"Les données fusionnées ont été sauvegardées dans {output_file}")
```

```python
# Charger le modèle de langue française de spaCy
nlp = spacy.load("fr_core_news_sm")

# Chemin vers le fichier CSV
input_file = '../data/histoires_resumes.csv'
output_file = '../data/histoires_resumes_annotes.csv'

# Lire le fichier CSV
df = pd.read_csv(input_file)

# Fonction pour effectuer l'analyse morphosyntaxique
def analyze_text(text):
    doc = nlp(text)
    tokens = []
    for token in doc:
        tokens.append(f"{token.text}/{token.pos_}")
    return " ".join(tokens)

# Appliquer l'analyse morphosyntaxique aux colonnes 'histoire' et 'resumes'
df['h_annot'] = df['histoire'].apply(analyze_text)
df['r_annot'] = df['resumes'].apply(analyze_text)

# Sélectionner uniquement les colonnes 'id', 'histoire_annotee' et 'resumes_annotes'
df = df[['id', 'h_annot', 'r_annot']]

# Enregistrer les résultats dans un nouveau fichier CSV
df.to_csv(output_file, index=False, encoding='utf-8')

print(f"Les données annotées ont été sauvegardées dans {output_file}")
```
